## Output the plot in a HTML page 

```python
from bokeh.plotting import figure, show
from bokeh.sampledata.iris import flowers
from bokeh.models import ColumnDataSource

# Create a ColumnDataSource from the iris dataset
source = ColumnDataSource(flowers)

# Create a scatter plot of sepal_width vs sepal_length with the iris dataset
p = figure(title="Iris Morphology")
p.xaxis.axis_label = "Sepal Length"
p.yaxis.axis_label = "Sepal Width"
p.scatter("sepal_length", "sepal_width", source=source) 

# Show the plot
show(p)
```

## Output the plot within Jupyter Notebook

To display Bokeh plots inline in Jupyter notebook, output_notebook() function from bokeh.io is used 

plot will display inline in the next notebook output cell following call to show() function  

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.sampledata.iris import flowers

# Enable the Bokeh output in the Jupyter notebook
output_notebook()

# Create a scatter plot of sepal_width vs sepal_length with the iris dataset
p = figure(title="Iris Morphology")
p.xaxis.axis_label = "Sepal Length"
p.yaxis.axis_label = "Sepal Width"
p.scatter("sepal_length", "sepal_width", source=flowers)

# Show the plot
show(p)
```
